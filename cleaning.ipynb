```python
formats = (
    '%d-%m-%Y', '%d-%m-%y',
    '%Y-%m-%d', '%Y/%m/%d',
    '%d/%m/%Y'
)

def parse_date_safe(x):
    if pd.isna(x):
        return pd.NaT
    x = str(x)
    for f in formats:
        try:
            return pd.to_datetime(x, format=f)
        except ValueError:
            continue
    return pd.NaT

order_dt['order_date'] = order_dt['order_date'].apply(parse_date_safe)
```

```python
order_dt['order_date'] = order_dt['order_date'].dt.normalize()
```

```python
order_dt.head(10)
```

```python
order_dt['city']= order_dt['city'].str.title()
```

```python
order_dt['payment_method']= order_dt['payment_method'].str.title()
```

```python
order_dt['payment_method'] = order_dt['payment_method'].str.replace('Upi','UPI')
```

```python
del_per_dt = pd.read_csv("raw data/delivery_performance.csv")
del_per_dt.head(10)
```

```python
from word2number import w2n

def quantity_to_number(x):
    try:
        return int(x)
    except:
        try:
            return w2n.word_to_num(str(x).lower())
        except:
            return None

del_per_dt['actual_time_mins'] = del_per_dt['actual_time_mins'].apply(quantity_to_number)
```

```python
del_per_dt.dtypes
```

```python
del_per_dt['actual_time_mins'] = pd.to_numeric(del_per_dt['actual_time_mins'])
```

```python
del_per_dt['actual_time_mins']= del_per_dt['actual_time_mins'].fillna(round(del_per_dt['actual_time_mins'].mean()))
```

```python
del_per_dt['actual_time_mins'] = pd.to_numeric(del_per_dt['actual_time_mins']).astype('Int32')
```

```python
del_per_dt.dtypes
```

```python
del_per_dt.isna().sum()
```

```python
del_per_dt['delivery_partner_rating'] = del_per_dt['delivery_partner_rating'].fillna(round(del_per_dt['delivery_partner_rating'].mean(),1))
```

```python
del_per_dt.isna().sum()
```

```python
del_per_dt.to_csv("cleaned data/delivery_performance.csv",index=False)
```

```python
cust_fb_dt['rating']=cust_fb_dt['rating'].str.replace('four','4').str.replace('five','5')
```

```python
def rating_category(rating):
    if rating < 3:
        return 'Not Good'
    elif  rating < 4:
        return 'Good'
    else:
        return 'Very Good'
    
cust_fb_dt['rating_category'] = cust_fb_dt['rating'].apply(rating_category)
```
